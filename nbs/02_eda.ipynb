# 2. EDA

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 250)
pd.set_option('display.max_colwidth', None)

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import seaborn as sns
```

```python
df = pd.read_parquet('../data/79AyushHH.parquet')

print(df.shape)
df.head()
```

```python
ax = df.hist(figsize=(30, 25), bins=20);
plt.suptitle('Histograms of the features in ~/data/79AyushHH.parquet',fontsize=20)
plt.subplots_adjust(top=0.95)
```

```python
filepath = '../data/79Ayushper.parquet'
df = pd.read_parquet(filepath)

print(df.shape)
df.head()
```

```python
ax = df.hist(figsize=(45, 35), bins=20);
plt.suptitle(f'Histograms of the features in {filepath}',fontsize=20)
plt.subplots_adjust(top=0.95)
```
