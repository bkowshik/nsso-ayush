# Notebook 3. DOT

Load data into `public.nsso_ayush` table in Postgres database to run tests using [DOT](https://github.com/datakind/Data-Observation-Toolkit/).

```python
import os

import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 250)
pd.set_option('display.max_colwidth', None)

from sqlalchemy import create_engine
```

```python
# Read dataset from inside the data directory.
df = pd.read_parquet("../data/79AyushHH.parquet")
print(df.shape)
df.head()
```

```python
# Read password to connect to Postgres from the environment.
password = os.getenv("POSTGRES_PASSWORD")
engine = create_engine(f"postgresql://postgres:{password}@localhost:5433/dot_db")
```

```python
# Instead of the entire dataset, loading only a subset into the database table.
df.sample(10000).to_sql("nsso_ayush", engine, schema="public", if_exists="replace", index=False, method="multi")
```
